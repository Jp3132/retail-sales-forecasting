```python
from dotenv import load_dotenv
import requests
from fredapi import Fred
import pandas as pd
import os
load_dotenv()
```

```python
api_key = os.getenv("FRED_API_KEY")
```

```python
 fred = Fred(api_key=api_key)
 
```

```python
data = fred.get_series("RSXFSN")
df = pd.DataFrame(data,columns=['sales_amount'])
df['sales_amount'] = df['sales_amount'].map(lambda sales_amount: sales_amount/1000)
df.index.freq = 'MS'
df
```

```python
sales_tuples = [(index.date(),float(sales)) for index,sales in df.itertuples()]
```
